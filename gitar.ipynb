Open the website and click on play 

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options 
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from PIL import Image
import re
import time

PATH = "C:\Program Files (x86)\chromedriver.exe"
driver = webdriver.Chrome(service=Service(PATH))
options = webdriver.ChromeOptions()
options.add_extension('C:\\Users\\\Downloads\\Video-Speed-Controllervideo-manager.crx')
driver = webdriver.Chrome(options=options) 
driver.get("https://www.youtube.com/watch?v=3u692mRAZhc&ab_channel=SkyGuitar")

# Find the element using XPath
video_name = driver.find_element("xpath","//*[@id='title']/h1/yt-formatted-string")

# Extract the text from the element
text = video_name.text
clean_text = re.sub(r'[^a-zA-Z]', '', text)
print(clean_text)
# Wait for the video player to load
time.sleep(5)

# Play the video
play_button = WebDriverWait(driver, 3).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'button.ytp-play-button')))
play_button.click()
# Find the video element
body = driver.find_element(by=By.TAG_NAME,value="body")
body.send_keys("m")
for i in range(25):
    body.send_keys("d")
```

crop and save the images 

```python
# Find the video element using XPath
import time
video_element = driver.find_element("xpath", "//*[@id='movie_player']/div[1]/video")
location = video_element.location
size =video_element.size

for i in range(100):
    driver.save_screenshot(str(i)+".png")
    x = location['x'] 
    print('X = ' + str(x))
    y = location['y']
    print('Y = ' + str(y))
    w = size['width'] 
    print('width =' +str(w))
    h = size['height']
    print('height =' +str(h))
    width = x + w
    height = y + h
    im = Image.open(str(i)+".png")
    im_edit = im.crop((int(x), int(y) + 260, int(width), int(height)))
    # x =24 , y =340 
    im_check =im.crop((int(x) + 10 ,int(y) + 280 , int(x) + 70 ,int(y) + 260 + 70))
    im_check.save(str(i) + 'check.png')
    im_edit.save(str(i)+'edit.png')
    time.sleep(2)
```

```python
counter = [0]*100
print(counter)
```

```python
import cv2
import numpy as np
# Load the images

for i in range(0,99):
    a = cv2.imread(str(i)+"check.png")
    b = cv2.imread(str(i+1) + "check.png")
    a_gray = cv2.cvtColor(a, cv2.COLOR_BGR2GRAY)
    b_gray = cv2.cvtColor(b, cv2.COLOR_BGR2GRAY)
    total_pixels = a_gray.shape[0] * a_gray.shape[1]
    a_white_pixels = cv2.countNonZero(a_gray)
    # Check if images are successfully loaded
    if a is None or b is None:
        print("Failed to load one or more images.")
    else:
        # Resize the images to the same dimensions
        resized_a = cv2.resize(a, (b.shape[1], b.shape[0]))

        # Compare images by computing the absolute difference
        difference = cv2.absdiff(resized_a, b)

        # Calculate the mean of the absolute difference
        mean_diff = np.mean(difference)

        if mean_diff < 0.7 or a_white_pixels < total_pixels/2:
            print("Pictures are the same")
        else:
            counter[i] += 1
            print("Pictures are different")
            
```
